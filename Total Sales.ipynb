```python
import pandas as pd
# need state 1
week_to_month = {1: 1, 2: 1, 3: 1, 4: 1,
                 5: 1, 6: 2, 7: 2, 8: 2,
                 9: 2, 10: 3, 11: 3, 12: 3,
                 13: 3, 14: 4, 15: 4, 16: 4,
                 17: 4, 18: 4, 19: 5, 20: 5,
                 21: 5, 22: 5, 23: 6, 24: 6,
                 25: 6, 26: 6, 27: 7, 28: 7,
                 29: 7, 30: 7, 31: 7, 32: 8,
                 33: 8, 34: 8, 35: 8, 36: 9,
                 37: 9, 38: 9, 39: 9, 40: 10,
                 41: 10, 42: 10, 43: 10, 44: 10,
                 45: 11, 46: 11, 47: 11, 48: 11,
                 49: 12, 50: 12, 51: 12, 52: 12}
df_sales = pd.read_csv('Total Sales - UTSC Lecture.csv')
#df_sales.head()
df_sales["Fiscal Month"] = df_sales['Fiscal Week'].map(week_to_month)
dict = {"Need State 1": "Total Sales"}
out=df_sales[["Fiscal Year", "Fiscal Week", "Fiscal Month", "Need State 1"]]
out[out["Fiscal Year"] == 2021]
out=out.rename(columns=dict)
out
```
